бидејќи сме сигурни дека во тој случај мин никогаш нема да ја избере таа патека, оваа исто така важи и за мин играчот односно
имаме пресликување mirroing(со бета). Имаме огромна преднсот затоа што се намалува процесорската моќ како и брзината на извршување на самиот
алгоритам, додека за недостатоци генерално нема, единствено во поглед на тоа што нема да извршуваме до крај одредено дрво,

ChatGPT верзија подобрена
Алфа-бета поткаструвањето е оптимизација на minimax алгоритмот
кој служи за намалување на бројот на јазли што треба да се 
разгледуваат при пребарување на играчко дрво. Наместо да се 
евалуира целото дрво, алгоритмот ги отфрла(поткаструва) оние
гранки за кои однапред е јасно дека нема да влијаат на 
конечната одлука.

Алгоритмот користи две променливи:
    - Алфа - најдобрата(највисоката) вредност што досега ја пронашол MAX
    играчот
    - Бета - најдобрата(најниската) вредност што досега ја пронашоч MIN играчот

При пребарување 
    - MAX се обидува да ја зголеми алфа
    - MIN се обидува да ја намали бета

Во моментот кога ќе се исполни условот алфа > бета, 
понатамошното разгледување на таа гранка се прекинува
се врши поткастрување, бидејќи е сигурно дека спротивниот
играч никогаш нема да ја избере таа патека. Овој принцип важи симетрично
за двата играчи(MAX и MIN), што претстаува т.н. mirroring на процесот.


Предности:
    Значително се намалува бројот на евалуирани јазли
    Се зголемува брзината на извршување
    Се намалува процесорската и мемориската побарувачка
    Овозможува пребарување на подлабоки нивоа во исто време

Недостатоци:
    Ефикасноста силно зависи од редолседот на разлледување на потезите
    Во практични игри како шах одредени ризични или жртвувачки потези
    можат рано да изгледаат лошо и да бидат поткастрени, иако
    подоцна би донеле голема предност
    Пример за ваков стил е шахистот Михаил Таљ, кој често жртвувал
    фигури за долгорочна иницијатива.




1.2. **(20 поени)** Кога играч во дадена игра е ограничен со време за секој потег, едно решение е да поставиме фиксна максимална длабочина, `MAX_DEPTH`, во самата имплементацијата на минимакс алгоритамот. Меѓутоа, ова решение не е баш најдобро бидејќи:
1. на почетокот на играта лесно може да истече времето додека минимакс пребарува.
2. на крајот на играта останува многу време, време кое може да се потроши ефективно така што ќе алгоритамот ќе пребарува подлабоко.

Предложете како да се решат овие проблеми, па имплементирајте ги во пајтон. Слободно можете да го имплементирате решението врз веќе постоечки код лабораториските вежби, на пример играта Поврзи 4.

```python
E = EMPTY = '·'
R = 'R'
Y = 'Y'
inf = float('inf')
MAX_DEPTH = 3
```

```python
def expand_state(state, player):
    for x in range(7):
        y = sum([1 if row[x] == EMPTY else 0 for row in state]) - 1
        if y == -1:
            continue
        new_state = list(map(list, state))
        new_state[y][x] = 'R' if player == 'MAX' else 'Y'
        yield new_state, x
```

```python
def evaluate_state(state, player):
    score = 0
    ROWS = 6
    COLS = 7
    EMPTY = '·'
    symbol = 'R' if player == 'MAX' else 'Y'
    opp_symbol = 'Y' if player == 'MAX' else 'R'

    # Check if a player has won (4 in a row)
    def check_win(sym):
        # Check horizontal locations for win
        for r in range(ROWS):
            for c in range(COLS - 3):
                if state[r][c] == sym and state[r][c+1] == sym and state[r][c+2] == sym and state[r][c+3] == sym:
                    return True
        # Check vertical locations for win
        for c in range(COLS):
            for r in range(ROWS - 3):
                if state[r][c] == sym and state[r+1][c] == sym and state[r+2][c] == sym and state[r+3][c] == sym:
                    return True
        # Check diagonally (positive slope)
        for r in range(ROWS - 3):
            for c in range(COLS - 3):
                if state[r][c] == sym and state[r+1][c+1] == sym and state[r+2][c+2] == sym and state[r+3][c+3] == sym:
                    return True
        # Check diagonally (negative slope)
        for r in range(ROWS - 3):
            for c in range(3, COLS):
                if state[r][c] == sym and state[r+1][c-1] == sym and state[r+2][c-2] == sym and state[r+3][c-3] == sym:
                    return True
        return False

    # Check for immediate wins/losses
    if check_win(symbol):
        return float('inf')    
    if check_win(opp_symbol):
        return float('-inf')

    # Check for a draw
    if all(cell != EMPTY for row in state for cell in row):
        return 0

    # Calculate score for a single window of 4 cells
    def score_window_calc(window, current_score):
        if window.count(symbol) == 4:
            current_score += 1000 # Should ideally be infinity, but for window scoring it's a high number
        elif window.count(symbol) == 3 and window.count(EMPTY) == 1:
            current_score += 100
        elif window.count(symbol) == 2 and window.count(EMPTY) == 2:
            current_score += 20
        elif window.count(opp_symbol) == 3 and window.count(EMPTY) == 1:
            current_score -= 80
        elif window.count(opp_symbol) == 2 and window.count(EMPTY) == 2:
            current_score -= 5
        return current_score
    
    # Calculate score across all windows
    # Horizontal
    for r in range(ROWS):
        for c in range(COLS - 3):
            window = [state[r][c + i] for i in range(4)]
            score = score_window_calc(window, score)
    # Vertical
    for c in range(COLS):
        for r in range(ROWS - 3):
            window = [state[r + i][c] for i in range(4)]
            score = score_window_calc(window, score)
    # Diagonal \
    for r in range(ROWS - 3):
        for c in range(COLS - 3):
            window = [state[r + i][c + i] for i in range(4)]
            score = score_window_calc(window, score)
    # Diagonal /
    for r in range(3, ROWS):
        for c in range(COLS - 3):
            window = [state[r - i][c + i] for i in range(4)]
            score = score_window_calc(window, score)

    # Add center column priority
    center_col = COLS // 2
    center_cells = [state[i][center_col] for i in range(ROWS)]
    score += center_cells.count(symbol)
    score -= center_cells.count(opp_symbol)

    return score

table = [['R', 'Y', 'Y', 'Y', 'Y', '·', '·'], 
         ['·', '·', '·', '·', '·', '·', '·'], 
         ['·', '·', 'Y', '·', '·', '·', '·'], 
         ['R', '·', '·', 'R', '·', 'Y', '·'], 
         ['·', '·', '·', '·', '·', '·', '·'], 
         ['·', '·', '·', '·', '·', '·', '·']]

print(evaluate_state(table, "MIN"))
```

import time
from math import inf

class TimeUp(Exception):
    pass

def minimax_timed(node, player, max_depth, alpha=-inf, beta=inf, depth=0, deadline=None):
    # time check
    if deadline is not None and time.perf_counter() >= deadline:
        raise TimeUp

    best_value = inf if player == 'MIN' else -inf
    best_move = None

    other_player = 'MIN' if player == 'MAX' else 'MAX'

    # Ако сме на лимит-длабочина, евалуирај и стопирај
    if depth >= max_depth:
        return evaluate_state(node, player), None

    # Ако имаш terminal check (win/draw), додади тука:
    # if is_terminal(node):
    #     return evaluate_state(node, player), None

    for child, move in expand_state(node, player):
        value, _ = minimax_timed(child, other_player, max_depth,
                                 alpha, beta, depth + 1, deadline)

        if player == 'MIN':
            if value < best_value:
                best_value = value
                best_move = move
            beta = min(beta, best_value)
            if beta <= alpha:
                break  # prune

        else:  # MAX
            if value > best_value:
                best_value = value
                best_move = move
            alpha = max(alpha, best_value)
            if alpha >= beta:
                break  # prune

    return best_value, best_move

def choose_move_iterative(node, player, time_limit_sec = 2.0, depth_cap = 50):
    start = time.perf_counter()
    safety = 0.01 # buffer
    deadline = start + max(0.0, time_limit_sec - safety)

    best_move = None
    best_value = None
    last_completed_depth = 0

    for d in range(1, depth_cap + 1):
        try:
            value, move = minimax_timed(node, player, max_depth = d, deadline=deadline)

            best_value, best_move = value, move
            last_completed_depth = d
        except TimeUp:
            break

    return best_move, best_value, last_completed_depth

table = [['R', 'Y', 'Y', 'Y', 'Y', '·', '·'], 
         ['·', '·', '·', '·', '·', '·', '·'], 
         ['·', '·', 'Y', '·', '·', '·', '·'], 
         ['R', '·', '·', 'R', '·', 'Y', '·'], 
         ['·', '·', '·', '·', '·', '·', '·'], 
         ['·', '·', '·', '·', '·', '·', '·']]

move, value, depth_done = choose_move_iterative(table, "MIN", depth_cap = 20)
print("Best move:", move, "value:", value, "depth:", depth_done)

1.3. **(5 поени)** Зошто ни треба функција за проценка на победникот (евристика) кај имплементацијата на минимакс кај шахот?

Па во секој вид на игра која ке ја играме со минимакс, алгоритмот
односно каде што ке развиваме вештачка интелигенција
која ке ја игра ни е потребно евристика, односно
проценка на состојбата до која сме стигнале. Принципот на работа
на шахот како игра со два играчи не е ништо поразлично од тоа, 
но кај шахот функцијата за евристика е многу сложена
затоа што имаме голем број на променливи кои фигурираат во истата.
Но пример за добро разивен chess engione кој се базира на min_max
алгоритмот е StockFish каде што имаме развиено сложена евристика.


Функцијата за проценка(евристика) кај минимакс е неопходна бидејќи не е возможно да се пребара целото дрво на играта поради огромниот број на можни состојби. Затоа алгоритмот мора да запре на одредена длабочина и со евристика да процени кој играч е во подобра позиција. Без ваква функција минимакс не би бил применлив во реално време. Пример за шаховски енџин со развиена евристика е Stockfish.



1.4. **(10 поени)** Еден начин на кој луѓето размислуваат додека играат шах е да се фокусираат на веројатните / логичните / реалните состојби, а да ги занемарат нелогичните потези (жртвување кралица за да се постигне предност подоцна), иако постои можност овие потези да донесат предност во остатокот од играта. Минимакс алгоритамот, каков што го знаеме од предавања и вежби, не размислува вака. Објаснете што би промениле за минимакс да ја добие способноста на луѓето како што е опишана тука. Дали всушност ќе добиеме подобрување или ова, сепак, е уназадување на алгоритамот?

Првенствено ви споменал дека тоа ке биде уназадување на алгоримот
во секоја смисла. Компјутерот има голема процесорска моќ и за разлика
од човекот може да развие одредена позиција на таблата дури до крајот.
Човекот е ограничен е е можно да ги развие позициите до тој степен
така да не би имал способност да предвиди одреден потег како би влијаел
затоа и ги занемарува(но има и играчи велемајстори како Михаил Таљ што
жртуваат многу фигури за да дојдат подоцна до предност).

Сепак оваа размислување може да се најде во примена во оптимизација
на алгоритмот при што во евристика состојбите жртвување ке бидат
обележани со знаменце за да не се разгледуваат.


За минимакс алгоритмот да размислува како човек, би требало да се воведе селективно пребарување, односно да се разгледуваат само "логични" потези, а да се игнорираат 
нетипичните(како жртвување фигури). Сепак, оваа претставува уназадување на алгоритмот, бидејќи минимакс ја губи својата целосност и може да пропушти потези кои на подолг рок носат голема преднсот. За разлика од човекот, компјутерот нема вакви ограничувања и може да ги разгледува и овие потези. Ваквиот пристап е оправдан само како оптимизација(на пр. преку евристика или редослед на потези), но не и како целосно отфралање на одредени состојби.
