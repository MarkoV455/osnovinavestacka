Шест луѓе треба да ја преминат реката. Дедо, баба, татко, мајка, момче и девојче.
   * Само мажите знаат да веслаат.
   * Чамецот може да издржи највеќе двајца.
   * Дедото не сака да биде придружуван во чамецот.
   * Девојчето може да ја премине реката единствено со татко и.

```python
grandpa_index, grandma_index, dad_index, mum_index, boy_index, girl_index, boat_index = range(7)
```

```python
initial_state = (0, 0, 0, 0, 0, 0, 0)
goal_state = (1, 1, 1, 1, 1, 1, 1)

new_state = list(initial_state)
new_state
tuple(new_state)
```

```python
def one_men_goes(state, index):
    new_state = list(state)
    new_state[index] = 1
    new_state[boat_index] = 1
    return tuple(new_state)

one_men_goes(initial_state, 0)
```

```python
def one_men_comes(state, index):
    new_state = list(state)
    new_state[index] = 0
    new_state[boat_index] = 0
    return tuple(new_state)

one_men_comes(goal_state, 0)
```

```python
def two_people_go(state, index_first, index_second):
    new_state = list(state)
    new_state[index_first] = 1
    new_state[index_second] = 1
    new_state[boat_index] = 1
    return tuple(new_state)

two_people_go(initial_state, 1, 2)
```

```python
def two_people_come(state, index_first, index_second):
    new_state = list(state)
    new_state[index_first] = 0
    new_state[index_second] = 0
    new_state[boat_index] = 0
    return tuple(new_state)

two_people_come(goal_state, 1, 2)
```

```python
def expand_state(state):
    new_states = []

    # Ako brodot e na 0 -tata strana od rekata
    if state[boat_index] == 0:
        # Proveruvame za dedoto na koja strana
        if state[grandpa_index] == 0:
            new_state = one_men_goes(state, grandpa_index)
            new_states.append(new_state)

        if state[dad_index] == 0 and state[girl_index] == 0:
            new_states.append(two_people_go(state, dad_index, girl_index))
        
        for men_index in [dad_index, boy_index]:
            if state[men_index] == 0:
                new_states.append(one_men_goes(state, men_index))
                for other_index in [dad_index, boy_index, grandma_index, mum_index]:
                    if other_index != men_index and state[other_index] == 0:
                        new_states.append(two_people_go(state, men_index, other_index))
       
        
    if state[boat_index] == 1:
        # Proveruvame za dedoto na koja strana
        if state[grandpa_index] == 1:
            new_state = one_men_comes(state, grandpa_index)
            new_states.append(new_state)

        if state[dad_index] == 1 and state[girl_index] == 1:
            new_states.append(two_people_come(state, dad_index, girl_index))
        
        for men_index in [dad_index, boy_index]:
            if state[men_index] == 1: 
                new_states.append(one_men_comes(state, men_index))
                for other_index in [dad_index, boy_index, grandma_index, mum_index]:
                    if other_index != men_index and state[other_index] == 1:
                        new_states.append(two_people_come(state, men_index, other_index))
            
    return new_states

expand_state(initial_state)
expand_state(goal_state)
```

```python
from collections import deque
def search(initial_state, alg):
    visited = {initial_state}
    states_queue = deque([[initial_state]])
    while states_queue:
        vertex_list = states_queue.popleft()
        state_to_expand = vertex_list[-1]
        for next_state in expand_state(state_to_expand):
            if next_state not in visited:
                if next_state == goal_state:
                    return vertex_list + [next_state]
                visited.add(next_state)
                if alg == 'dfs':
                    states_queue.appendleft(vertex_list + [next_state])
                elif alg == 'bfs':
                    states_queue.append(vertex_list + [next_state])
    return None
    
```

```python
def one_men_goes(state, index):
    new_state = list(state)
    new_state[index] = 1
    new_state[boat_index] = 1
    return tuple(new_state)

one_men_goes(initial_state, 0)
```

```python
def one_men_comes(state, index):
    new_state = list(state)
    new_state[index] = 0
    new_state[boat_index] = 0
    return tuple(new_state)

one_men_comes(goal_state, 0)
```

```python
def two_people_go(state, index_first, index_second):
    new_state = list(state)
    new_state[index_first] = 1
    new_state[index_second] = 1
    new_state[boat_index] = 1
    return tuple(new_state)

two_people_go(initial_state, 1, 2)
```

```python
def two_people_come(state, index_first, index_second):
    new_state = list(state)
    new_state[index_first] = 0
    new_state[index_second] = 0
    new_state[boat_index] = 0
    return tuple(new_state)

two_people_come(goal_state, 1, 2)
```

```python
def expand_state(state):
    states = []

    if state[boat_index] == 0:
        # 1 човек оди (0->1)
        for index in [dad_1_index, dad_2_index, boy_1_index, boy_2_index]:
            if state[index] == 0:
                states.append(one_men_goes(state, index))

        # 2 момчиња одат заедно
        if state[boy_1_index] == 0 and state[boy_2_index] == 0:
            states.append(two_people_go(state, boy_1_index, boy_2_index))

    else:  # boat_index == 1
        # 1 човек се враќа (1->0)
        for index in [dad_1_index, dad_2_index, boy_1_index, boy_2_index]:
            if state[index] == 1:
                states.append(one_men_comes(state, index))

        # 2 момчиња се враќаат заедно
        if state[boy_1_index] == 1 and state[boy_2_index] == 1:
            states.append(two_people_come(state, boy_1_index, boy_2_index))

    return states

print(expand_state((1, 0, 1, 0, 1)))
```

```python
from collections import deque
def search(initial_state, alg):
    visited = {initial_state}
    states_queue = deque([[initial_state]])
    while states_queue:
        vertex_list = states_queue.popleft()
        state_to_expand = vertex_list[-1]
        for next_state in expand_state(state_to_expand):
            if next_state not in visited:
                if next_state == goal_state:
                    return vertex_list + [next_state]
                visited.add(next_state)
                if alg == 'dfs':
                    states_queue.appendleft(vertex_list + [next_state])
                elif alg == 'bfs':
                    states_queue.append(vertex_list + [next_state])
    return None
    
```
