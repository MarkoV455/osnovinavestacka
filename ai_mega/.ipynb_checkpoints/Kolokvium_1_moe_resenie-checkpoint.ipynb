## Задача 1 - Роботска Платформа

Разгледуваме роботска платформа од 16 плочки, распоредени како матрица 4х4. Вашата работна група треба под секоја плочка треба да намести сензор за детекција на објектите кои ќе бидат поставувани врз плочките. Ако се постави сензор под секоја плочка, се јавува проблем и сензорите не работат како што треба. По одредено истражување, вашата работна група сватила дека проблемот се јавува кога ќе има два сензори во ист ред, или пак два сензори во иста колона. Исто така вашата група открила дека два исти сензори, но од различен производител, не си пречат еден на друг кога ќе се во ист ред или колона. Пред да потрошете пари за нови сензори, треба да се реши проблемот: `Под која плочка каков сензор да се постави?`. Моментално располагате со неограничен број сензори од 4 различни производители, A, B, C, и D.

Еве едно можно решение за роботска платформа со 3х3 плочки.
$$ \left[ {\begin{array}{ccc} A & B & C \\ C & A & B \\ B & C & A \\ \end{array} } \right] $$

Одговори:

1) Почетната состојба ке ни биде непополнета табела, односно непотполента роботска платформа.
    Ке формираме торка од торки(односно матрица 4x4). Крајна состојба ке ни е потполнетата торка со сензори.

2) Разгранувањето ке се одвива на начин што ке итерираме низ самата состојба, ке го најдеме следното празно место
каде што може да поставиме сензор, на тоа место(таа позиција) ке изитерираме низ сите можни сензори
и ке провериме дали може да го поставиме тој сензор. Ке формираме нова променлива односно нова состојба
со сензорот на празната позиција и ке го додадеме во листата со состојби.

3) Одбирам да го користам алгоритмот со пребарување по длабочина. Овој алгоритам е најсоодветен во
овој случај од следните причини:
    - Немаме одредена цена за секој сензор(Затоа не го избираме A* или Djikstra)
    - Битно е само крајното решение(Не е битно најдоброто затоа не го избирам BFS)
    - Најбрзо решение ке се добие со овој алгоритам

```python
N = 5
sensors = []
for i in range(N):
    sensors.append(i)
```

```python
# Deklariranje na pocetna sostojba
initial_state = tuple((None,) * N for _ in range(N))
initial_state
```

```python
def check_end(state):
    return all(value != None for row in state for value in row)

check_end(initial_state)
```

```python
initial_state_list = list(list(row) for row in initial_state)
def find_empty_slot(state):
    for row_i, row in enumerate(state):
        for col_i, value in enumerate(row):
            if value == None:
                return row_i, col_i

print(find_empty_slot(initial_state_list))
```

```python
def expand_state(state):
    new_states = []
    empty_slot_row, empty_slot_col = find_empty_slot(state)
    for sensor in sensors:
        if any(state[empty_slot_row][i] == sensor for i in range(len(state))):
            continue
        if any(state[i][empty_slot_col] == sensor for i in range(len(state))):
            continue
        new_state = list(list(row) for row in state)
        new_state[empty_slot_row][empty_slot_col] = sensor
        new_states.append(tuple(tuple(row) for row in new_state))

    return new_states

expand_state(initial_state)
```

```python
from collections import deque
def search(initial_state, alg):
    visited = {initial_state}
    states_queue = deque([initial_state])
    while states_queue:
        state_to_expand = states_queue.popleft()
        for next_state in expand_state(state_to_expand):
            if next_state not in visited:
                if check_end(next_state):
                    return next_state
                visited.add(next_state)
                if alg == 'dfs':
                    states_queue.appendleft(next_state)
                elif alg == 'bfs':
                    states_queue.append(next_state)
```
