```python
class Graph():
    def __init__(self):
        self.graph_dict = {}

    def add_vertex(self, vertex):
        if vertex not in self.graph_dict:
            self.graph_dict[vertex] = []

    def vertices(self):
        return list(self.graph_dict.keys())

    def add_edge(self, edge, add_reversed=True):
        vertex1, vertex2 = edge
        self.graph_dict[vertex1].append(vertex2)
        if add_reversed:
            self.graph_dict[vertex2].append(vertex1)

    def edges(self):
        edges = []
        for vertex in self.graph_dict:
            for neighbour in self.graph_dict[vertex]:
                edges.append((vertex, neighbour))
        return edges

    def neighbours(self, vertex):
        return self.graph_dict[vertex]

    def remove_vertex(self, vertex_to_remove):
        del self.graph_dict[vertex_to_remove]
        for vertex in self.vertices():
            if vertex_to_remove in self.graph_dict[vertex]:
                self.graph_dict[vertex].remove(vertex_to_remove)

    def remove_edge(self, edge_to_remove, remove_reversed=True):
        vertex1, vertex2 = edge_to_remove
        if vertex2 in self.graph_dict[vertex1]:
            self.graph_dict[vertex1].remove(vertex2)
        if remove_reversed:
            if vertex1 in self.graph_dict[vertex2]:
                self.graph_dict[vertex2].remove(vertex1)

    def isolated_vertices(self):
        isolated_vertices = []
        for vertex in self.graph_dict:
            if not self.graph_dict[vertex]:
                isolated_vertices.append(vertex)
        return isolated_vertices
```

```python
from collections import deque

def BFS(graph, starting_vertex, goal_vertex):
    if starting_vertex == goal_vertex:
        return []

    visited = {starting_vertex}
    queue = deque([[starting_vertex]])
    while queue:
        vertex_list = queue.popleft()
        vertex_to_expand = vertex_list[-1]
        for neighbour in graph.neighbours(vertex_to_expand):
            if neighbour not in visited:
                visited.add(neighbour)
                if neighbour == goal_vertex:
                    return vertex_list + [neighbour]
                queue.append(vertex_list + [neighbour])

print(BFS(new_graph, "S", "E"))
```

```python
from collections import deque

def DFS(graph, starting_vertex, goal_vertex):
    if starting_vertex == goal_vertex:
        return []

    visited = {starting_vertex}
    queue = deque([[starting_vertex]])

    while queue:
        vertex_list = queue.popleft()
        vertex_to_expand = vertex_list[-1]

        for neighbour in graph.neighbours(vertex_to_expand):
            if neighbour not in visited:
                visited.add(neighbour)
                if neighbour == goal_vertex:
                    return vertex_list + [neighbour]
                queue.appendleft(vertex_list + [neighbour])

print(DFS(new_graph, "S", "E"))
```

```python
def transfer_to_other_glass(source, sink, sink_capacity):
    return max(source - (sink_capacity - sink), 0), min(source + sink, sink_capacity)
```

```python
def expand_state(state):
    capacities = (5, 8, 10)
    glass_0, glass_1, glass_2 = state

    new_states = []

    # Top up the glasses
    if glass_0 < capacities[0]:
        new_states.append((capacities[0], glass_1, glass_2))
    if glass_1 < capacities[1]:
        new_states.append((glass_0, capacities[1], glass_2))
    if glass_2 < capacities[2]:
        new_states.append((glass_0, glass_1, capacities[2]))

    # Empty the glasses
    if glass_0 > 0:
        new_states.append((0, glass_1, glass_2))
    if glass_1 > 0:
        new_states.append((glass_0, 0, glass_2))
    if glass_2 > 0:
        new_states.append((glass_0, glass_1, 0))

    # Transfer helper (optional: avoid duplicates/no-op moves)
    def add_if_changed(s):
        if s != state:
            new_states.append(s)

    # From 1 to 2 (0 -> 1)
    new_source, new_sink = transfer_to_other_glass(glass_0, glass_1, capacities[1])
    add_if_changed((new_source, new_sink, glass_2))

    # From 2 to 1 (1 -> 0)
    new_source, new_sink = transfer_to_other_glass(glass_1, glass_0, capacities[0])
    add_if_changed((new_sink, new_source, glass_2))

    # From 1 to 3 (0 -> 2)
    new_source, new_sink = transfer_to_other_glass(glass_0, glass_2, capacities[2])
    add_if_changed((new_source, glass_1, new_sink))

    # From 3 to 1 (2 -> 0)
    new_source, new_sink = transfer_to_other_glass(glass_2, glass_0, capacities[0])
    add_if_changed((new_sink, glass_1, new_source))

    # From 2 to 3 (1 -> 2)
    new_source, new_sink = transfer_to_other_glass(glass_1, glass_2, capacities[2])
    add_if_changed((glass_0, new_source, new_sink))

    # From 3 to 2 (2 -> 1)  fixed
    new_source, new_sink = transfer_to_other_glass(glass_2, glass_1, capacities[1])
    add_if_changed((glass_0, new_sink, new_source))

    return new_states
```

```python
def check_valid(state, row, col):
    # Only check rows from 0 up to (but not including) the current row
    for r in range(row):
        existing_col = state[r]
        
        # If we hit a None somehow, skip it (safety check)
        if existing_col is None:
            continue
            
        # 1. Vertical Check
        if existing_col == col:
            return False
            
        # 2. Diagonal Check
        if abs(existing_col - col) == abs(r - row):
            return False
            
    return True
```

```python
def expand_state_queens(state):
    new_states = []
    try:
        free_row = state.index(None)
    except ValueError:
        return []
    for i in range(N):
        if check_valid(state, free_row, i):
            new_state = list(state)
            new_state[free_row] = i
            new_states.append(new_state)
    return new_states
```

```python
from collections import deque

def search_path_queens(initial_state, alg='dfs'):
    # Using a deque as a flexible stack/queue
    states_queue = deque([initial_state])
    
    while states_queue:
        # Both BFS and DFS take from the left in this specific setup
        state_to_expand = states_queue.popleft()
        
        for next_state in expand_state_queens(state_to_expand):
            # GOAL TEST: If no 'None' remains, we've placed 8 queens!
            if next_state.count(None) == 0:
                return next_state
            
            if alg == 'dfs':
                # LIFO (Last-In, First-Out) -> Stack behavior
                states_queue.appendleft(next_state)
            elif alg == 'bfs':
                # FIFO (First-In, First-Out) -> Queue behavior
                states_queue.append(next_state)
                
    return None
```

```python
def find_empty_space(state):
    for r_idx, row in enumerate(state):
        for c_idx, val in enumerate(row):
            if val == 0:
                return (r_idx, c_idx)

find_empty_space(g)
```

```python
def expand_state_sudoku(state):
    new_states = []

    spot = find_empty_space(state)
    if spot is None:
        return []  # no empty cells -> solved, no expansions

    row, col = spot

    for value in range(1, 10):
        # row check
        if value in state[row]:
            continue

        # column check
        if any(state[r][col] == value for r in range(9)):
            continue

        # 3x3 square check
        r0 = (row // 3) * 3
        c0 = (col // 3) * 3
        if any(state[r][c] == value for r in range(r0, r0 + 3) for c in range(c0, c0 + 3)):
            continue

        # make a deep copy and place value
        new_state = [list(r) for r in state]
        new_state[row][col] = value

        # keep it immutable/hashable (tuple of tuples)
        new_states.append(tuple(tuple(r) for r in new_state))

    return new_states
```

```python
from collections import deque

def search(initial_state, alg, expand_state, check_end):
    if check_end(initial_state):
        return initial_state

    visited = {initial_state}
    states_queue = deque([initial_state])

    while states_queue:
        state_to_expand = states_queue.popleft()

        for next_state in expand_state(state_to_expand):
            if next_state in visited:
                continue

            if check_end(next_state):
                return next_state

            visited.add(next_state)

            if alg == "dfs":
                states_queue.appendleft(next_state)  # stack-like
            elif alg == "bfs":
                states_queue.append(next_state)      # queue-like
            else:
                raise ValueError("alg must be 'dfs' or 'bfs'")

    return None
```

```python
class WeightedGraph:
    def __init__(self):
        self.graph_dict = {}

    def add_vertex(self, vertex):
        if vertex == None:
            return

        if vertex not in self.graph_dict.keys():
            self.graph_dict[vertex] = {}
        

    def vertices(self):
        return list(self.graph_dict.keys())

    def add_edge(self, edge, add_reversed = True):
    
        vertex1, vertex2, weight = edge
        self.graph_dict[vertex1][vertex2] = weight
        if add_reversed:
            self.graph_dict[vertex2][vertex1] = weight

    def edges(self):
        edges = []
        for vertex in self.graph_dict.keys():
            for neighbour, weight in self.graph_dict[vertex].items():
                edges.append((vertex, neighbour, weight))

        return edges
    
    def remove_edge(self, edge, remove_reversed = True):
        vertex1, vertex2, weight = edge

        if vertex1 in self.graph_dict and vertex2 in self.graph_dict[vertex1]:
            del self.graph_dict[vertex1][vertex2]

        if remove_reversed and vertex2 in self.graph_dict and vertex1 in self.graph_dict[vertex2]:
            del self.graph_dict[vertex2][vertex1]

    def remove_vertex(self, vertex):
        if vertex not in self.graph_dict:
            return

        # Remove all edges pointing TO this vertex
        for v in self.graph_dict:
            if vertex in self.graph_dict[v]:
                del self.graph_dict[v][vertex]

        # Remove the vertex itself
        del self.graph_dict[vertex]

                
    def neighbours(self, vertex):
       return list(self.graph_dict[vertex].items())
```

```python
import heapq
def uniform_coast_search(graph, starting_vertex, goal_vertex):
    if starting_vertex == goal_vertex:
        return []

    expanded = set()

    queue = [(0, [starting_vertex])]
    heapq.heapify(queue)

    while queue:
        
        weight, vertex_list = heapq.heappop(queue)
        vertex_to_expand = vertex_list[-1]

        if vertex_to_expand in expanded:
            continue
        if vertex_to_expand == goal_vertex:
            return vertex_list, weight

        for neighbour, new_weight in graph.neighbours(vertex_to_expand):
            if neighbour not in expanded:
                heapq.heappush(queue, (weight + new_weight, vertex_list + [neighbour]))
        
        expanded.add(vertex_to_expand)
```

```python
import heapq
def a_star_search(graph, starting_vertex, goal_vertex, heuristic_function, alpha = 1):
    #Dokolku se raboti za istiot jazel nema potreba algoritmot shto da istrazuva
    
    if starting_vertex == goal_vertex:
        return

    expanded = set()
    # Formirame lista koja se sostoi od torka kade gi smestuvame a_star_weight, current weight vo edna torka
    # I edna lista kade gi smestuvame jazlite
    queue = [((0, 0),[starting_vertex])]

    heapq.heapify(queue)

    while queue:
        weight_tuple, vertex_list = heapq.heappop(queue)

        current_a_star_weight, current_path_weight = weight_tuple
        vertex_to_expand = vertex_list[-1]

        if vertex_to_expand == goal_vertex:
            return vertex_list, current_path_weight
    
        if vertex_to_expand  in expanded:
            continue

        for neighbour, new_weight in graph.neighbours(vertex_to_expand):
            
            heuristic = heuristic_function(neighbour, goal_vertex)
            path_weight = current_path_weight + new_weight

            a_star_weight = path_weight + alpha*heuristic
            
            
            heapq.heappush(queue, ((a_star_weight, path_weight), vertex_list + [neighbour]))
            
        expanded.add(vertex_to_expand)

    
```

```python
df = pd.read_csv("../code/data/roads/coordinates.csv")

def euclidean_distance(vertex1, vertex2):
    vertex1 = int(vertex1.split(' - ')[0])
    vertex2 = int(vertex2.split(' - ')[0])
    vertex1 = df.query(f" == {vertex1}")
    vertex2 = df.query(f" == {vertex2}")
    import math
    delta_x_squared = math.pow(float(vertex1.X) - float(vertex2.X), 2)
    delta_y_squared = math.pow(float(vertex1.Y) - float(vertex2.Y), 2)
    return math.sqrt(delta_x_squared + delta_y_squared)


euclidean_distance('364 - ', '378 -   1')
```
